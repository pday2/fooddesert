```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
# education
edu_file = './education/ACSST5Y2020.S1501_data_with_overlays_2022-04-27T105238.csv'
# Read column names from file
cols = list(pd.read_csv(edu_file, nrows =1))
#print(cols)

# Use list comprehension to remove the unwanted column in **usecol**
# columns with last alpha = M are margin of error data
df = pd.read_csv(edu_file, usecols =[col for col in cols if col[-1]!='M'], na_values='(X)')
```

```python
# axis=0 -> rows, axis=1 -> cols
edu = df.dropna(axis=1)
```
